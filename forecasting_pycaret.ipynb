```python
import warnings
warnings.filterwarnings('ignore')
import math
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

# from prophet import Prophet

from pycaret.time_series import *
```

```python
countries = {"Germany": "de", "Spain": "es", "France": "fr", "India": "in", "Italy": "it", "Russia": "ru", "Turkey": "tr", "United States": "us"}
data_formats = ["daily_data", "cumulative_data"]
country_list = list(countries.keys())
country_iso_code_list = list(countries.values())
data_type_list = ["confirmed_cases", "recovered_cases", "death_cases", "active_cases"]
```

```python
def get_data(data_format, country_code, data_type, start_date, end_date):
    df = pd.read_csv("./data/" + data_format + "/" + country_code + "_" + data_type + ".csv", index_col=False)
    df["ds"] = pd.to_datetime(df["ds"])
    period = (df['ds'] >= start_date) & (df['ds'] <= end_date)
    df = df.loc[period]
    df.set_index("ds", inplace=True)
    return df
```

```python
end_date = "2020-05-20"

# US
us_start_date = "2020-01-27"
us_confirmed = get_data(data_format=data_formats[0], country_code=country_iso_code_list[7], data_type=data_type_list[0], start_date=us_start_date, end_date=end_date)
us_death = get_data(data_format=data_formats[0], country_code=country_iso_code_list[7], data_type=data_type_list[2], start_date=us_start_date, end_date=end_date)
us_confirmed_cumulative = get_data(data_format=data_formats[1], country_code=country_iso_code_list[7], data_type=data_type_list[2], start_date=us_start_date, end_date=end_date)

# ES
es_start_date = "2020-02-06"

# IT
it_start_date = "2020-02-05"

# FR
fr_start_date = "2020-01-29"

# DE
de_start_date = "2020-02-01"

# RU
ru_start_date = "2020-02-05"

# TR
tr_start_date = "2020-03-16"

# IN
in_start_date = "2020-02-04"

```
